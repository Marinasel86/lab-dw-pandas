```python
# mean, median, mode, standard deviation, and quartiles to understand the central tendency and distribution of the data for numerical columns.
```

```python
numerical_columns = file1_df.select_dtypes(include=['int64', 'float64']).columns
mean = file1_df[numerical_columns].mean()
print("mean:")
print(mean)
median = file1_df[numerical_columns].median()
print("median:")
print(median)
mode = file1_df[numerical_columns].mode().iloc[0]
print("mode:")
print(mode)
std = file1_df[numerical_columns].std()
print("standard deviation:")
print(std)
q1 = file1_df[numerical_columns].quantile(0.25)
q2 = file1_df[numerical_columns].quantile(0.50)
q3 = file1_df[numerical_columns].quantile(0.75)
print("25 procentages are:")
print(q1)
print("50 procentages are:")
print(q2)
print("75 procentages are:")
print(q3)
```

```python
# Create a Pandas Series with customer locations and their frequencies
customer_location_series = file1_df['ST'].value_counts()

# 5 less common locations
top_5_less_common_locations = customer_location_series.nsmallest(5)

print("Top 5 Less Common Customer Locations:")
print(top_5_less_common_locations)
```
